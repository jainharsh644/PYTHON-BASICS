```python
#INTRO TO DECORATORS : used to enhance the functionallity of other functions
# @ used for decorator 

def dec_func(any_func):
    def wrapper_function():
        print("This is awesome function")
        any_func()
    return wrapper_function

@dec_func              #syntactic sugar 
def func1():
    print("This is function 1")
func1()

@dec_func
def func2():
    print("This is function 2")
func2()
#func1 = dec_func(func1)
#func1()
```

```python
#DECORATORS PART 2 :
def dec_func(any_func):
    def wrapper_function(*args, **kwargs):
        print("This is awesome function")
        return any_func(*args, **kwargs)     
    return wrapper_function
@dec_func
def func(a):
    print(f"This is function with arg {a} ")
func(5)
@dec_func
def add(a,b):
    return a+b
add(2,3)          #5 will be not printed as function is returning the value.
```

```python
#DECORATORS PART 3
from functools import wraps
def dec_func(any_func):
    @wraps(any_func)
    def wrapper_function(*args, **kwargs):
        """this is wrapper function """
        print("This is awesome function")
        return any_func(*args, **kwargs)     
    return wrapper_function

@dec_func            #due to this were getting such output as it indirectly calls wrapper_func
def add(a,b):
    '''this is add function '''
    return a+b        
print(add.__doc__)   # output - this is wrapper function
                     # how to solve this? "" by importing wraps from functools "" and 
                     # decorate the wrap func before defining any inner function  
    
print(add.__name__)
```

```python
#DECORATORS PRACTICE :

from functools import wraps

def print_function_data (function):
    @wraps(function)
    def wrap_func(*args, **kwargs):
        print(f"You are calling {function.__name__} function")
        print(f"{function.__doc__}")
        return function(*args, **kwargs)
    return wrap_func

@print_function_data        
def add(a,b):
    '''This is a function takes 2 numbers as arguments and return their sum'''
    return a+b

print(add(4,5))

# OUTPUT WILL BE - you are calling add function.
                 # This is a function takes 2 numbers as arguments and return their sum  
```

```python
#EXERCISE - 1
from functools import wraps
import time 

def calculate_time(function):
    @wraps(function)
    def wrap_func(*args, **kwargs):
        print(f'Executing ... {function.__name__}')
        t1 = time.time()
        return_value = function(*args, **kwargs)
        t2 = time.time()
        total = t2-t1
        print(f"The time taken is {total} secs")
        return return_value
    return wrap_func

@calculate_time
def sq(a):
    return [i**2 for i in range(1, a+1)]

sq(500)   
```

```python
#DECORATOR PRACTICE 2 :

#only int allow decorator :


from functools import wraps
def only_int_allow(function):
    @wraps(function)
    def wrap_func(*args, **kwargs):
        data_types = []
        for arg in args:
            data_types.append(type(arg) == int)
        if all(data_types):
            return function(*args, **kwargs)
        else:
            print("Invalid datatype")
    return wrap_func    


@only_int_allow       
def add_all(*args):
    total = 0
    for i in args:
        total += i
    return total

print(add_all(2,3,4,5,[4,5,6]))   #the problem is we cant pass the list to this function TO SOLVE THIS 
                          # WE USE " DECORATORS "
```

```python
#short the above code :

from functools import wraps
def only_int_allow(function):
    @wraps(function)
    def wrap_func(*args, **kwargs):
        if all([type(arg)==int for arg in args]):
            return function(*args, **kwargs)
        print("Invalid Argument")
    return wrap_func

@only_int_allow       
def add_all(*args):
    total = 0
    for i in args:
        total += i
    return total

print(add_all(2,3,4,5,[4,5,6]))
```

```python
#DECORATORS WITH ARGUMENTS : DEEPER DIVE IN DECORATORS :

from functools import wraps

def only_datatype_allow(data_type):
    def decorator(function):
        @wraps(function)
        def wrap_func(*args, **kwargs):
            if all([type(arg)==data_type for arg in args]):
                return function(*args, **kwargs)
            print("Invalid Argument")
        return wrap_func
    return decorator

@only_datatype_allow(str)
def string_join(*args):
    string = ''
    for i in args:
        string += i
    return string

string_join('harsh', 'jai', 'n')        
            
```
